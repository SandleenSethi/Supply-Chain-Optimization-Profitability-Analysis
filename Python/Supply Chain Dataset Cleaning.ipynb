                        Data Cleaning

```python
drop_cols = [
    'Customer Email',
    'Customer Password',
    'Customer Fname',
    'Customer Lname',
    'Customer Street',
    'Product Image',
    'Product Description'
]

df = df.drop(columns=drop_cols)
df.shape
```

```python
# Fix all object (text) columns
object_cols = df.select_dtypes(include='object').columns

for col in object_cols:
    df[col] = df[col].astype(str).str.strip()

df[object_cols].head()
```

                    Feature Engineering

```python
df['order date (DateOrders)'] = pd.to_datetime(df['order date (DateOrders)'])
df['shipping date (DateOrders)'] = pd.to_datetime(df['shipping date (DateOrders)'])
```

```python
df['actual_shipping_days'] = (
    df['shipping date (DateOrders)'] - df['order date (DateOrders)']
).dt.days
```

```python
df['delivery_delay'] = (
    df['Days for shipping (real)'] - df['Days for shipment (scheduled)']
)
```

```python
df['is_delayed'] = df['delivery_delay'].apply(lambda x: 1 if x > 0 else 0)
```

                      KPI creation

```python
total_sales = df['Sales'].sum()
total_sales
```

```python
total_profit = df['Order Profit Per Order'].sum()
total_profit
```

```python
profit_margin = (total_profit / total_sales) * 100
profit_margin
```

```python
total_orders = df['Order Id'].nunique()
total_orders
```

```python
late_delivery_pct = df['is_delayed'].mean() * 100
late_delivery_pct
```

```python
avg_delivery_time = df['actual_shipping_days'].mean()
avg_delivery_time
```
